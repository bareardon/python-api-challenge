```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from scipy.stats import linregress
from datetime import date 

# Import API key
#from api_keys import weather_api_key
my_api_source = open('/Users/bridg/OneDrive/Desktop/weather_api_key.txt')
weather_api_key = my_api_source.read()

# Import todays date in mm/dd/yyyy format
today = date.today()
today = today.strftime("%m/%d/%Y")

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(lat_range[0], lat_range[1], size=1500)
lngs = np.random.uniform(lng_range[0], lng_range[1], size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
# Save config information 
url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + weather_api_key
   

```

```python
# Create empty lists to populate weather data 
city_id = []
city_name = []
cloudiness = []
country = []
humidity = []
latitude = []
longitude = []
max_temperature = []
wind_speed = []
date = []
counter = 0
sets = 1
error = 0 

```

```python
# Loop through each city
for city in cities:
    
     # Create search query, make request and store in json
    query = url + "&q=" + city
    response = requests.get(query)
    response_json = response.json()
   
    # Create loop to fill lists 
    try:

        city_name.append(response_json['name'])
        country.append(response_json['sys']['country'])
        max_temperature.append(response_json ['main']['temp_max'])
        humidity.append(response_json['main']['humidity'])
        cloudiness.append(response_json['clouds']['all'])
        wind_speed.append(response_json['wind']['speed'])
        latitude.append(response_json['coord']['lat'])
        longitude.append(response_json['coord']['lon'])
        date.append(response_json['dt'])
        
        if counter >= 50: 
            counter = 1
            sets += 1
            time.sleep(1.00)
        else:
            counter += 1
            
        print(f"Processing Record {counter} of set {sets} | {city}")
       
       
    except:
        # Append null values
        print("City not found. Next...")
        
# Once search is complete, print 'Processing complete'
print(f"Processing complete")
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
weather_dict = {
    "city": city_name,
    "lat": latitude,
    "lon": longitude,
    "max temp": max_temperature,
    "humidity": humidity,
    "cloudiness": cloudiness,
    "wind speed": wind_speed,
    "country": country,
     "date": date}

city_data = pd.DataFrame(weather_dict)
city_data.head()

```

```python
# Export city data into a .csv
city_data.to_csv(output_data_file, index = False, header = True)
```

## Linear Regression

```python
# Find index for hemispheres
index_northern = city_data[city_data['lat']>=0].index
index_southern = city_data[city_data['lat']<=0].index

# Create data frames for hemispheres
northern_df = city_data.drop(index_southern)
southern_df = city_data.drop(index_northern)
```

####  Northern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
x_values = northern_df['lat']
y_values = northern_df['max temp']

(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))

plt.scatter(x_values, y_values, alpha=0.5, edgecolor='black')
plt.plot(x_values,regress_values,"r-")

plt.title(f'City Latitude vs. Max Temperature ({today})')
plt.xlabel('Latitude')
plt.ylabel('Max Temperature (F)')
plt.grid()
plt.savefig("output_data/Northern_LatitudeVsTemperaturePlot.png")
print(f"The r-value is: {rvalue**2}")
print(f"The line equation is: {line_eq}")
plt.show()
```

####  Southern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
x_values = southern_df['lat']
y_values = southern_df['max temp']

(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))

plt.scatter(x_values, y_values, alpha=0.5, edgecolor='black')
plt.plot(x_values,regress_values,"r-")

plt.title(f'City Latitude vs. Max Temperature ({today})')
plt.xlabel('Latitude')
plt.ylabel('Max Temperature (F)')
plt.grid()
plt.savefig("output_data/Southern_LatitudeVsTemperaturePlot.png")
print(f"The r-value is: {rvalue**2}")
print(f"The line equation is: {line_eq}")
plt.show()
```
